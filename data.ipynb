```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Earthquakes_database.csv")
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.drop(columns='Root Mean Square', inplace=True)
df
```

```python
df [['Month', 'Day', 'Year']] = df['Date'].str.split('/', n=2, expand=True)
df
```

```python
df = df.replace('NaN', '')
df = df.replace('', np.nan)
df
```

```python
bins = [5.49, 6.0, 7.0, 9.11] 
df['categories'] = pd.cut(df['Magnitude'], bins=bins, labels=['Low', 'Medium', 'High'], include_lowest=True)

df
```

```python
melted_data = pd.melt(df, id_vars=['Date', 'Time'], value_vars=['Latitude', 'Longitude', 'Depth', 'Magnitude'])
print(melted_data.head())
```

```python
pivot_table_data = pd.pivot_table(df, index='Year', values='Magnitude', aggfunc='mean')

print(pivot_table_data)
```

```python
# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Magnitude'], bins=15, kde=False)
plt.title('Frequency Distribution of Earthquake Magnitudes')
plt.xlabel('Magnitude')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
# Line chart
plt.figure(figsize=(10, 6))
magnitude_by_year = df.groupby('Year')['Magnitude'].mean()
magnitude_by_year.plot(kind='line')
plt.title('Trend of Earthquake Magnitudes Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Magnitude')
plt.grid(True)
plt.show()
```

```python
# Bar chart
plt.figure(figsize=(10, 6))
sns.histplot(df['Depth'], bins=20, kde=False)
plt.title('Distribution of Earthquake Depths')
plt.xlabel('Depth (km)')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
# Scatter chart
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Magnitude', y='Depth', data=df)
plt.title('Earthquake Magnitudes vs Depths')
plt.xlabel('Magnitude')
plt.ylabel('Depth (km)')
plt.grid(True)
plt.show()
```

```python
df['Date'] = df['Date'].astype(str)
df['Time'] = df['Time'].astype(str)

df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], errors='coerce')

df.set_index('Datetime', inplace=True)

df = df[df.index.notnull()]

df['Year'] = df.index.year
df['Month'] = df.index.month
df['Day'] = df.index.day
df['Day_of_Week'] = df.index.dayofweek

monthly_earthquakes = df.resample('M').size()

print(df.head())
print(monthly_earthquakes)
```
